Выбор архитектуры связан со временем обучения:хочется, чтобы модели обучались меньше часа на моем ноутбуке.

````python
def epoch(net, data):
    loss_total = 0
    
    for x_batch, y_batch in get_batches(data, batch_size):
        
        net.zeroGradParameters()

        predictions = net.forward(x_batch)
        loss = criterion.forward(predictions, y_batch)
        
        if net.training:
            dp = criterion.backward(predictions, y_batch)
            net.backward(x_batch, dp)
            optimizer(net.getParameters(), 
                      net.getGradParameters(), 
                      optimizer_config,
                      optimizer_state)      
    
        loss_total += loss * x_batch.shape[0]
    return loss_total / data[0].shape[0]
````

````python
class BestModel(nn.Module):
    def __init__(self):
        super(self.__class__, self).__init__()
        
        self.model = nn.Sequential(
            nn.Linear(784, 50),
            nn.BatchNorm1d(50),
            nn.LeakyReLU(),
            nn.Dropout(p=0.6),
            nn.Linear(50, 50),
            nn.LeakyReLU(),
            nn.Dropout(p=0.6),
            nn.Linear(50, 10),
            nn.LogSoftmax()
        )
    
    def forward(self, inp):
        out = self.model(inp)
        return out
````

````python
def epoch(model, data, training=True):
    model.train(training)
    
    criterion = nn.NLLLoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=1e-2)
    correct = 0.0;
    
    for x_batch, y_batch in get_batches(data, batch_size):
        
        x_batch = torch.FloatTensor(x_batch)
        y_batch = torch.FloatTensor(y_batch)
        
        predictions = model(x_batch)
        loss = criterion(predictions, y_batch.argmax(axis=1))
        
        if training:
            loss.backward()
            optimizer.step()
            optimizer.zero_grad()
            
        correct += (predictions.argmax(dim=1) == y_batch.argmax(dim=1)).sum()
    
    return correct / data[0].shape[0]
````
